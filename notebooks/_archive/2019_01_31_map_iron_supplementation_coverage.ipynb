```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import pandas as pd
# Connect Javascript to display the plots in the notebook
init_notebook_mode(connected=True)

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')

```

```python
data = [ dict(
        type = 'choropleth',
        locations = df['CODE'],
        z = df['GDP (BILLIONS)'],
        text = df['COUNTRY'],
        colorscale = [[0,"rgb(5, 10, 172)"],[1,"#ff00ff"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            # autotick = False,
            tickprefix = '$',
            title = 'GDP<br>Billions US$'),
      ) ]

layout = dict(
    title = '2014 Global GDP<br>Source:\
            <a href="https://www.cia.gov/library/publications/the-world-factbook/fields/2195.html">\
            CIA World Factbook</a>',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
iplot(fig)
```

```python
from src.data.download_DHS import load_country_codebook
country_codebook = load_country_codebook()[['CountryName', 'ISO3_CountryCode']]
```

```python
# AN_MIAM_W_90P (90 day iron supp)
# RH_ANCC_W_IRN (any iron supp)
df = query_dhs_api('AN_MIAM_W_90P')
# Keep most recent by country
df = df.sort_values(['CountryName','SurveyYear'], ascending=False).drop_duplicates('CountryName')
# Merge country ISO3
df = pd.merge(country_codebook, df)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
data = [dict(type='choropleth',
           locations=df['ISO3_CountryCode'], 
           z = df['Value'],
            text=df['CountryName'],
             # colorscale='Blackbody', #scale, Rainbow
             colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
            colorbar = {'title': 'Indicator'}
            )]
```
